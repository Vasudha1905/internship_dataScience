## Regression Plots

### Linear Regression
* linear regression computes the linear relationship between the dependent variable and one or more independent features by fitting a linear equation to observed data.
* evaluates whether a set of predictor variables do a good job in predicting an outcome (dependent) variable


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
sns.set(palette='deep',rc={"figure.figsize": (5, 4)})
```

#### LM plot vs Reg plot
 regplot is used for simple scatter plots with a regression line, ideal for quick analysis. lmplot offers more flexibility with faceting, allowing you to create complex visualizations by splitting data into subplots.

### LM Plot

```python
sns.set_style('whitegrid')
```

```python
# loading dataset
tips=sns.load_dataset('tips')
sns.lmplot(x='total_bill',y='tip',data=tips,palette='Blues')
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=tips,palette='deep')
```

changing markers

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, markers=["o", "x"], palette="Set1")
```

seperate plots

```python
sns.lmplot(x="total_bill", y="tip", row='sex',col="smoker", data=tips)
```

### Reg Plot

```python
sns.regplot(x="total_bill", y="tip", data=tips, color="g")
```

### Joint plot

```python
# Loading Built-in Dataset:
tips = sns.load_dataset("tips")
sns.jointplot(x="total_bill", y="tip", data=tips, color='b')
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips,kind="reg")
```

adding a specif point

```python
a=sns.jointplot(x="total_bill", y="tip", data=tips, color='b')
a.ax_joint.plot([0.8],[0.8],'bx',color='g')
a.ax_joint.text(0.8,0.8,"Random point")
```
