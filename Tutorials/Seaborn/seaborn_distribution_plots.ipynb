## Distribution Plots
* distplot()
* jointplot()
* pairplot()
* rugplot()
* kdeplot()



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings
%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
sns.set(rc={"figure.figsize": (3, 3)})
sns.set_style('ticks')
```

#### loading dataset

```python
tips = sns.load_dataset('tips')
```

#### distplot()

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

#### jointplot()

```python
# scatter
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter',edgecolor='k')
```

```python
#hex
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

#### pairplot()

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='viridis')
```

#### rugplot()
draw a dash mark for every point on a univariate distribution.

```python
sns.rugplot(tips['total_bill'])
```

#### kdeplot()

```python
sns.kdeplot(tips['total_bill'])
```
