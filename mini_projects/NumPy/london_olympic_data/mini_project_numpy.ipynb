## CASE STUDY - NUMPY

```python
import numpy as np
```

```python
country = np.array(['Great Britain','China','Russia','United States','Korea','Japan','Germany'])
gold_medal = np.array([29,38,24,46,13,7,11])
silver_medal = np.array([17,28,25,28,8,14,11])
bronze_medal = np.array([19,22,32,29,7,17,14])
```

### maximum number of gold medals earned

```python
gold_max=max(gold_medal)
print(gold_max)
```

```python
# to find index
gold_medal_max = gold_medal.argmax()
gold_medal_max
```

### country with highest gold medals 

```python
# index 3 in country array
country[gold_medal_max]
```

### countries with more than 20 gold medals

```python
country[gold_medal>20]
```

### total number of medals in the olympics

```python
np.sum(gold_medal + silver_medal + bronze_medal)
```

### how many gold medals does each country have?

```python
for i in range(len(country)):
    print(country[i], " has",gold_medal[i]," gold medals")
```

### total medals won by each country

```python
total_medals = gold_medal + silver_medal + bronze_medal
for i in range(len(country)):
    print(country[i], " has won",total_medals[i],"total medals")
```

### What is the average number of medals won per country?

```python
average_medal= np.average(total_medals)
print(average_medal.round())
```

### data visualisation

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(country,total_medals,width=0.4,)
plt.title("Total medals won by each country")
```

```python
x = np.arange(7) 

plt.bar(x-0.2, gold_medal, width=0.2, color='gold') 
plt.bar(x, silver_medal, width=0.2, color='silver') 
plt.bar(x+0.2, bronze_medal, width=0.2, color='peru') 
plt.title("Number of Medals Won by each Country")
plt.xticks(x, country) 
plt.xlabel("country") 
plt.ylabel("number of medals") 
plt.legend(["gold", "silver", "bronze"]) 

```

```python
from matplotlib import style
plt.pie(gold_medal,autopct='%1.0f%%')
plt.title("distribution of gold medals by country")
plt.legend(country,bbox_to_anchor=(1.35, 1.0))
plt.show()
```
