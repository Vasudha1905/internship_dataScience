## Area and Population Project
* In this project we use the size of points to indicate the area and populations of California cities.
* We will create a legend that specifies the scale of the sizes of the points

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### loading data and viewing

```python
data=pd.read_csv('california_cities.csv')
```

```python
data.head(10)
```

#### extracting the data we need

```python
# population
population=data['population_total']
```

```python
# area
area=data['area_total_km2']
```

```python
# latitude and longitude
latitude=data['latd']
longitude=data['longd']
```

#### visualisation

```python
plt.scatter(longitude,latitude,c=np.log10(population),cmap='viridis',s=area,alpha=0.5,linewidth=0)   # s-size, c-colour, alpha-transperency, linewidth-point outline
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.title('Area and Population of California Cities')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)   # colourbar limit

# creating label
for ar in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=ar, label=str(ar)+'km$^2$')   # c=k --> black
plt.legend(scatterpoints=1, labelspacing=1, title='City Areas')
plt.show()
```

#### showing elevation using colour scale

```python
elevation=data['elevation_ft']
```

```python
plt.scatter(latitude,longitude,cmap='viridis',c=np.log10(elevation),alpha=0.3)
plt.colorbar()
plt.clim(0,4)
```
